# 生成数据集

我们构造一个简单的人工训练数据集，它可以使我们能够直观比较学到的参数和真实的模型参数的区别。设训练数据集样本数为1000，输入个数（特征数）为2。给定随机生成的批量样本特征 X∈R1000×2X∈R 
1000×2
 ，我们使用线性回归模型真实权重 w=[2,−3.4]⊤w=[2,−3.4] 
⊤
  和偏差 b=4.2b=4.2，以及一个随机噪声项 ϵϵ 来生成标签

```python
num_inputs = 2
num_examples = 1000
true_w = [2, -3.4]
true_b = 4.2
features = torch.randn(num_examples, num_inputs,
                       dtype=torch.float32)
labels = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_b
# np.random.normal(0, 0.01, size=labels.size())
# 其中噪声项 ϵϵ 服从均值为0、标准差为0.01的正态分布。
```

```python
# 它每次返回batch_size（批量大小）个随机样本的特征和标签。
def data_iter(bacth_size, features, labels):
    num_examples = len(features)  # 1000
    indices = list(range(num_examples))
    # 将序列x洗净，就是重新排序indices
    random.shuffle(indices)  # 样本的读取顺序是随机的
    for i in range(0, num_examples, batch_size):
        j = torch.LongTensor(indices[i: min(i + batch_size, num_examples)])
        yield features.index_select(0, j), labels.index_select(0, j)
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, y)
    break
```

# 初始化模型参数

我们将权重初始化成均值为0、标准差为0.01的正态随机数，偏差则初始化成0。

```python
w = torch.tensor(np.random.normal(0, 0.01, (num_inputs, 1)), dtype=torch.float32)
b = torch.zeros(1, dtype=torch.float32)
```

```python
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

定义模型

```python
def linreg(X, w, b):
    return torch.mm(X, w) + b
```

定义损失函数

```python
def squared_loss(y_hat, y):
    return (y_hat - y.view(y_hat.size())) ** 2 / 2
```

定义优化算法

```python
def sgd(params, lr, batch_size):
    for param in params:
        param.data -= lr * param.grad / batch_size
```

训练模型

```python
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    # x和y分别是小批量样本的特征和标签
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y).sum()  # l是有关小批量X，y的损失
        l.backward()
        sgd([w, b], lr, batch_size)
        
        # 梯度清零
        w.grad.data.zero_()
        b.grad.data.zero_()
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().item()))
```
